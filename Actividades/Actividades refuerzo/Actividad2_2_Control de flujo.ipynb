**1.**- Escribe un código que pida por teclado una tempratura Celsius o Farenheit con el siguiente formato:
23C, 67F...

Se ha incluido la primera línea para pedir el texto por teclado.

```python
# escribe tu código
temperatura = input("Dime que temperatura hace")
if temperatura.endswith ("C"):
    print("Hay", temperatura, "grados")
elif temperatura.endswith ("F"):
    print("Hay", temperatura, "grados")
else:
    print("Hay", temperatura, "grados")
```
