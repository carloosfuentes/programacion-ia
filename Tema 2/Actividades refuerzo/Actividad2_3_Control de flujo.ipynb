**1.**-  Escribe in programa que pida un número al usuario y muestre su tabla de multiplicar. El número debe ser del 1 al 9. Debes hacerlo con una sentencia while.

```python
# escribe tu código
numero = int(input("Escriba el numero que usted quiera"))
i = 0

while i != 11:
    print(numero * i)
    i+=1
```

**3.**- 
Escribe un programa que pida números enteros de uno en uno, los meta en una lista y, cuando acabe (debes pedir introducir una letra en concreto para acabar) debe recorrer la lista y mostrar la suma de los pares y la suma de los impares.

```python
numeros = []

while True:
    dato = input("Escribe un número (o q para acabar): ")
    if dato == "q":
        break
    numeros.append(int(dato))

suma_pares = 0
suma_impares = 0

for n in numeros:
    if n % 2 == 0:
        suma_pares += n
    else:
        suma_impares += n

print("Suma de pares:", suma_pares)
print("Suma de impares:", suma_impares)
```
